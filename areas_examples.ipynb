```python
import osmnx as ox
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning) 
```

```python
#setup coordinate for norwich, and 1500m area.
point = (52.63, 1.29)
dist = 500 # this is half width.

#tags for later use.
tags =  {'building':True}
```

```python
#get the building foot prints for the city centre.
gdf = ox.geometries_from_point(point, tags, dist=dist)
gdf_proj = ox.project_gdf(gdf)

fig, ax = ox.plot_footprints(gdf_proj)
#get the sum of the area.
areas = gdf_proj.area
area = sum(areas)
```

```python
#total area covered by buildings in Norwich.
area
```

our area is a square, so the total area is $(dist*2)^2$ as it's a half width (radius)


```python
total_area = (dist*2)**2
proportion_covered = (area / total_area) * 100
f'{proportion_covered}% building coverage in city centre'
```
